#Spatial domain to frequency domain

Diffrence between Spatial Domain and Frequency Domain

##Spatial Domain 
In Spatial Domain,  Manipulation of images is done directly on pixels of an image.
![Spatial Domain](https://www.tutorialspoint.com/dip/images/frequencyDomain1.jpg)


##Frequency Domain 

Initially we transform the image to its corosponding frequency Distribution, frequncy distribution results are processed. the output of processing will not be an image, but a ***transformation***. After performing inverse transformation, it is converted into an image which is then viewed in spatial domain 

![Frequency Domain](https://www.tutorialspoint.com/dip/images/frequencyDomain2.jpg)

Term ***Transformation***
A signal can be converted from time domain into frequency domain using mathematical operators called Transforms.

Types of Transformers
*   Fourier Series
*   Fourier Transformation
*   Laplace Transform
*   Z transform

what are frequency Components?
any real image in spatial domain can be represented in a frequency domain. but question arises what do this frequencies actually mean

we will divide freqency components into two major components

*   **High frequency components**
    *   High freqency components correspond to edges in an image.
    
*   **Low frequeny components**
    *   Low frequency components in an image correspond to smooth regions.

```
import cv2
import numpy as np
import matplotlib.pyplot as plt


#Read the image as grayscale and convert it to "np.float32" datatype
image=cv2.imread('g.png',0).astype(np.float32)/255
```

##Applying Discrete Fourier Transform


we can implemetn fast fourier transform algorithm using ***cv2.dft*** function.
the functions support optional flags specifying whether the output should be real or complex(flags ***cv2.DFT_REAL_OUTPUT*** and ***cv2.DFT_COMPLEX_OUTPUT***, respectively) and whether output values should be scaled(using the ***cv2.DFT_SCALE*** flag)


```
fft=cv2.dft(image, flags=cv2.DFT_COMPLEX_OUTPUT)
```

## Visualizing the Image spectrum

the ***np.fft.fftshift*** fuunction shifts the frequency spectrum in such a way that the aplitude corrosponding to zero frequency beconmes located at the centre of the array, and it's easier to interpret






```

shifted=np.fft.fftshift(fft, axes=[0,1])
magnitude=cv2.magnitude(shifted[:,:,0],shifted[:,:,1])
magnitude=np.log(magnitude)
plt.axis('off')
plt.imshow(magnitude,cmap='gray')
plt.tight_layout()
plt.show()
```

## convert the image from the frequency spectrum back to spatial representation

```
restored=cv2.idft(fft, flags=cv2.DFT_SCALE | cv2.DFT_REAL_OUTPUT)

plt.imshow(restored)
```
